### 查看 CUDA + PyTorch 環境

````python
import torch

if torch.cuda.is_available():
    print("GPU 名稱：", torch.cuda.get_device_name(0))
    print("CUDA 版本：", torch.version.cuda)
    print("總記憶體：", round(torch.cuda.get_device_properties(0).total_memory / (1024**3), 2), "GB")
else:
    print("找不到 GPU，請確認是否已安裝 NVIDIA 驅動與 CUDA。")


print("\n--------------------------------------------------------\n")
print("系統是否能使用 CUDA：", torch.cuda.is_available())
print("取得 第 0 張 GPU 的名稱：", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "No GPU")

# 若無 GPU
# 找不到 GPU，請確認是否已安裝 NVIDIA 驅動與 CUDA。

# --------------------------------------------------------

# 系統是否能使用 CUDA： False
# 取得 第 0 張 GPU 的名稱： No GPU
````
