### 格式轉換
可以用 llama.cpp 執行的模型文件格式被稱為 gguf 格式。

#### 1. 先安裝相關套件：

```python
%%bash
pip install numpy sentencepiece
```

#### 2. 使用 llama.cpp 提供的 `convert_hf_to_gguf.py` 工具，將 Hugging Face 的模型轉換成 gguf 格式：
* 這樣就可以輕鬆做轉換了，而且速度相當快。
* 另外可以設定 `--outtype` 參數決定資料型態，這個階段支援 `f32`, `f16` 與 `q8_0`，推薦使用 `f16` 比較平衡一點。

```python
%%bash

cd llama.cpp
# # python convert_hf_to_gguf.py "~/.cache/huggingface/hub/models--meta-llama--Meta-Llama-3-8B-Instruct/" \
# #     --outfile llama3-8b-inst.gguf --vocab-type bpe
python3 convert_hf_to_gguf.py   ~/.cache/huggingface/hub/models--meta-llama--Meta-Llama-3-8B-Instruct/snapshots/5f0b02c75b57c5855da9ae460ce51323ea669d8a   --outfile llama-3-8b-inst.gguf
```

* 若出現 `convert_hf_to_gguf.py: error: unrecognized arguments: --vocab-type bpe` 錯誤
    * `--vocab-type` 是 舊版 GGUF 轉換流程中的參數，用於指定 `tokenizer` 格式（如 `bpe` 或 `sentencepiece`）。
    * 最新版 的 llama.cpp（和 `convert_hf_to_gguf.py`）會自動從 Hugging Face 模型中偵測 `tokenizer` 類型，不需要你手動指定。

* 注意：
    * 在對 Breeze 模型做轉換時，可能會遇到 `ValueError: Vocab size mismatch` 的問題，這是因為 Breeze 模型參數的字典維度與分詞器實際的字典數量不符合的關係，只要加上 `--pad-vocab` 參數來填充字典即可。
